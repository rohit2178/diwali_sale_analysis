```python
ax=sb.countplot(x='Gender',data=df,hue='Gender')
for bars in ax.containers:

    ax.bar_label(bars)
```

```python
gen_sale=df.groupby(['Gender'])['Amount'].sum().sort_values(ascending=False).reset_index()
sb.barplot(x='Gender',y='Amount',data = gen_sale,hue='Gender')



```

females have purchased more in diwali sale

```python
ax=sb.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
order = df['Age Group'].value_counts().index
ax = sb.countplot(x='Age Group', data=df, order=order)
ax.bar_label(ax.containers[0])
plt.show()
```

Most of the buyers from age roup of 26-35  females

```python
df.columns
```

```python
state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by=['Orders'],ascending=False).head(10)
plt.Figure(figsize=(40,10))
plt.xticks(rotation=90)
sb.barplot(x='State',y='Orders',data= state)

plt.show()
```

top three state on the basis of order we recieved from uttar pradesh,Maharashtra and karnataka

```python
amount=df.groupby(['State',],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sb.barplot(x='State',y='Amount',data=amount)
plt.xticks(rotation=90)
plt.show()
```

#top three state on the basis of revenue fo sale amount are UP,Maharshtra and karnataka

```python
df.Occupation.value_counts()
```

```python
ax=sb.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=90)
plt.show()
```

```python
## as we can see from above graph that top three occupation sector of them we recieved order from are IT sector ,Healthcare an aviation
```

```python
sales_sect=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x='Occupation',y='Amount',data=sales_sect)
plt.xticks(rotation=90)
plt.show()
```

# as we can see from above garph that top three sector we genrated our sale from IT sector,healthcare and aviation

```python
ax=sb.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=90)
plt.yticks([500,1000,1500,2000,2500,3000])
plt.show()
```

# as we can see from above graph our top three selling category are Clothing & apparel,food and electronics&gadgets

```python
cat_amt=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sb.barplot(x='Product_Category',y='Amount',data=cat_amt)
plt.xticks(rotation=90)
plt.show()
```
